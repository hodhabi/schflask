```python
import pandas as pd
from datetime import datetime
import calendar
import time

sch = pd.read_excel("schedule.xlsx")
```

```python
now = datetime.now()
current_time = now.strftime("%H.%M")

my_date = datetime.today()
d = calendar.day_name[my_date.weekday()]

if(d=="Monday"):
    dd = ("M")
elif (d=="Tuesday"):
    dd="T"
elif (d=="Wednesday"):
    dd = "W"
elif (d=="Thursday"):
    dd = "R"
elif (d=="Friday"):
    dd = "F"
elif (d=="Saturday"):
    dd = "S"
elif (d=="Sunday"):
    dd = "U"
```

```python
for i in range(10):
    s = sch[sch["Campus"]=="AA"]
    s = s[s["Course Day"].str.contains(dd)==True]
    snow = s[s["End Time"]>current_time]
    print(snow[["Course",'Course Title', "Start Time","End Time","Course Day"]])
    time.sleep(1)
    
```
